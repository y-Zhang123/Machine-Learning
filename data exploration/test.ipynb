```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X, y, random_state=1)
from sklearn.tree import DecisionTreeRegressor
melb_model=DecisionTreeRegressor(random_state=1)
melb_model.fit(train_X, train_y)
pred_y = melb_model.predict(test_X)
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(test_y, pred_y)
print("Mean Absolute Error:", mae)
comparison = pd.DataFrame({'Predicted': pred_y, 'Actual': test_y})
comparison.head()
```

```python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=1)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return mae
```

```python
mae = []
for max_leaf_nodes in [5, 50, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, test_X, train_y, test_y)
    print("Max leaf nodes: %d  \t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
    mae.append(my_mae)
```

```python
best_mae = min(mae)
best_index = mae.index(best_mae)
best_max_leaf_nodes = [5, 50, 500, 5000][best_index]
print("Best max_leaf_nodes:", best_max_leaf_nodes)
```

```python
final_model = DecisionTreeRegressor(max_leaf_nodes=best_max_leaf_nodes, random_state=1)
final_model.fit(X, y)
predicted_prices = final_model.predict(X)
print(predicted_prices) 
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

forest_model = RandomForestRegressor(random_state=1)
forest_model.fit(train_X, train_y)
forest_pred = forest_model.predict(test_X)
forest_mae = mean_absolute_error(test_y, forest_pred)
print("Random Forest MAE:", forest_mae)
```
